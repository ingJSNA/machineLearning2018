## 2. Combine all the data in one DataFrame

```python
import pandas as pd
import glob

profitability = pd.DataFrame()

# Go to download path and read all files
retval = os.getcwd()
os.chdir(download_path)
for file in glob.glob("reporteRentabilidades*.xls"):
    print(file)
    df = pd.read_excel(file)
    profitability = profitability.append(df)
    
os.chdir(retval)

profitability.describe(include="all")
```

## 3. Preprocessing

```python
profitability = profitability.drop_duplicates()

# Drop columns
profitability = profitability. drop(columns=['Tipo Entidad'], errors='ignore')

# Convert dates
profitability['Fecha corte'] = pd.to_datetime(profitability['Fecha corte'],infer_datetime_format=False)
profitability = profitability.sort_values(by=['Fecha corte']) 

profitability.to_excel(download_path+"/profitability.xls", index=False)

profitability.describe(include="all")
```
